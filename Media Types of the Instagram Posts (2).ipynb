```python
# Using InstaLoader to retrieve the media types of each post publisehd by PS from the 6th of Januray to the 9th of March
# creating a CSV file named media_types_of_PS with the columns 'Media Type' and 'Post Date'

import instaloader
import pandas as pd
from datetime import datetime, timedelta

# Create an instance of Instaloader
L = instaloader.Instaloader()

# Define the username
username = 'partido_socialista'

# Get the profile metadata
profile = instaloader.Profile.from_username(L.context, username)

# Define the start and end dates of the time period
start_date = datetime(2024, 1, 6)  
end_date = datetime(2024, 3, 9)   

# Create an empty list to store the data
data = []

# Iterate over the recent posts within the defined time period
try:
    for post in profile.get_posts():
        # Check if the post is within the specified time period
        if start_date <= post.date <= end_date:
            # Get the media type of the post
            media_type = post.typename
            # Get the date of the post
            post_date = post.date

            # Append data to the list
            data.append({'Media Type': media_type, 'Post Date': post_date})
        else:
            # If the post date is beyond the end date, stop the iteration
            if post.date < start_date:
                break
        
except instaloader.exceptions.ProfileNotExistsException:
    print(f"The profile '{username}' does not exist.")

# Convert the list of dictionaries to a dataframe
df = pd.DataFrame(data)

# Save it to CSV
df.to_csv('media_types_of_PS.csv', index=False)
```
