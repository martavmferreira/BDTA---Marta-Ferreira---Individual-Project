```python
# Plotting the line graph of the vote itention by party

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("vote_intention_by_party.csv", sep=";")

df["Votes"] = df["Votes"].str.replace(",", ".").astype(float)  # turn votes into floats

plt.figure(figsize=(12, 8))

for party in df["Party"].unique():
        party_data = df[df["Party"] == party]
        plt.plot(party_data["Time Period"], party_data["Votes"], marker='o', label=party)  # plot for each party the 
                                                                             # number of votes by each time period
plt.xticks(rotation=45)
plt.title('Vote Intention for Portuguese General Elections 2024')
plt.xlabel('Time Period')
plt.ylabel('Vote Intention (%)')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotting the line graph again without the parties 'PSD' and 'CDS'

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("vote_intention_by_party.csv", sep=";")

df["Votes"] = df["Votes"].str.replace(",", ".").astype(float)

# List of parties to exclude
exclude_parties = ['PSD', 'CDS']

plt.figure(figsize=(12, 8))

# Added the part to exclude the parties in the exclude_parties list
for party in df["Party"].unique():
    if party not in exclude_parties:
        party_data = df[df["Party"] == party]
        plt.plot(party_data["Time Period"], party_data["Votes"], marker='o', label=party)

plt.xticks(rotation=45)
plt.title('Vote Intention for Portuguese General Elections 2024')
plt.xlabel('Time Period')
plt.ylabel('Vote Intention (%)')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotting the line graph again without the time periods of Nov 14 to Nov 17 and Dec 18 to Dec 21 
# (when PPD/PSD-CDS-PP.PPM is zero)

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("vote_intention_by_party.csv", sep=";")

df["Votes"] = df["Votes"].str.replace(",", ".").astype(float)

# Drop rows with the specified time periods
drop_time_periods = ['Nov 14 to Nov 17, 2023', 'Dec 18 to Dec 21, 2023']
df = df[~df["Time Period"].isin(drop_time_periods)]

exclude_parties = ['PSD', 'CDS']

plt.figure(figsize=(12, 8))

for party in df["Party"].unique():
    if party not in exclude_parties:
        party_data = df[df["Party"] == party]
        plt.plot(party_data["Time Period"], party_data["Votes"], marker='o', label=party)

plt.xticks(rotation=45)
plt.title('Vote Intention for Portuguese General Elections 2024')
plt.xlabel('Time Period')
plt.ylabel('Vote Intention (%)')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
