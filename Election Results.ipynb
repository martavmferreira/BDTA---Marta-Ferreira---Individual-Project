```python
import pandas as pd
voting_results_df = pd.read_excel('election_results.xlsx')
voting_results_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sort the dataframe by the 'Votes' column in descending order
# but move 'Nulls' and 'Blanks' to the end
voting_results_df = pd.concat([
    voting_results_df[~voting_results_df['Political Party'].isin(['Nulls', 'Blanks'])].sort_values(by='Votes', ascending=False),
    voting_results_df[voting_results_df['Political Party'].isin(['Nulls', 'Blanks'])]
])

plt.figure(figsize=(16, 6))
bars = plt.bar(voting_results_df['Political Party'], voting_results_df['Votes'], color='skyblue')

plt.xlabel('Political Party')
plt.ylabel('Total Votes')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability
plt.title('Total Number of Votes for Each Political Party')

# Set y-axis limits manually
plt.ylim(0, 2000000) 

# Display the values on top of the bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(int(bar.get_height())), 
             ha='center', va='bottom')

plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
