```python
# Retrieval of the number of likes of PS from the 6th of January to the 9th of March

import instaloader
import pandas as pd

L = instaloader.Instaloader()

account_username = "partido_socialista"

# Define the start and end dates of the time period (in YYYY-MM-DD format)
start_date = "2024-01-06"
end_date = "2024-03-09"

# Retrieve the profile of the Instagram account
profile = instaloader.Profile.from_username(L.context, account_username)

# Initialize an empty list to store the likes
likes_list_PS = []

# Iterate through each post of the account
for post in profile.get_posts():
    # Check if the post date is within the specified time period
    if start_date <= post.date.strftime("%Y-%m-%d") <= end_date:
        likes_list_PS.append(post.likes)

# Print the number of likes for each post made during the specified time period
for i, likes in enumerate(likes_list_PS, start=1):
    print(f"Post {i}: {likes} likes")

df = pd.DataFrame({"Post Number": range(1, len(likes_list_PS)+1), "Likes": likes_list_PS})

# Save it to a CSV file
df.to_csv("likes_PS.csv", index=False)
```

```python
# Calculate the total number of likes received
total_likes_PS = sum(likes_list_PS)

print(f"Total likes received: {total_likes_PS}")
```

```python
# Create a new CSV file with all the likes dataframes merged so that it has the list of likes instead of the
# total like in total_likes_all_parties

import pandas as pd

# Load the likes datasets for each political party
df_likes_PS = pd.read_csv('likes_PS.csv')
df_likes_AD = pd.read_csv('likes_AD.csv')
df_likes_CH = pd.read_csv('likes_CH.csv')
df_likes_IL = pd.read_csv('likes_IL.csv')
df_likes_BE = pd.read_csv('likes_B.E..csv')
df_likes_PCP = pd.read_csv('likes_PCP.csv')
df_likes_L = pd.read_csv('likes_L.csv')
df_likes_PAN = pd.read_csv('likes_PAN.csv')

# Add a column for the political party
df_likes_PS['Political Party'] = 'PS'
df_likes_AD['Political Party'] = 'PPD/PSD.CDS-PP.PPM'
df_likes_CH['Political Party'] = 'CH'
df_likes_IL['Political Party'] = 'IL'
df_likes_BE['Political Party'] = 'B.E.'
df_likes_PCP['Political Party'] = 'PCP'
df_likes_L['Political Party'] = 'L'
df_likes_PAN['Political Party'] = 'PAN'

# Merge all likes datasets into one
df_all_likes = pd.concat([df_likes_PS, df_likes_AD, df_likes_CH, df_likes_IL, df_likes_BE, df_likes_PCP, df_likes_L, df_likes_PAN])

# Save the merged dataset to a new CSV file
df_all_likes.to_csv('all_parties_likes.csv', index=False)
```
