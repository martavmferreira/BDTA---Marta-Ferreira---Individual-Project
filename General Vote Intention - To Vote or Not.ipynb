```python
# Plot the bar graph of the general vote intention

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('intention_to_vote_or_not_vote.xlsx')

pastel_color = sns.color_palette('pastel')[3]

plt.figure(figsize=(8, 6))
plt.bar(df['Intention'], df['Respondents (in %)'], color=pastel_color)
plt.xlabel('Intention')
plt.ylabel('Amount of Respondents (in %)')
plt.title('Voting Intention')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Displaying the values on top of the bars
for i, val in enumerate(df['Respondents (in %)']):
    plt.text(i, val, f"{val}%", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
