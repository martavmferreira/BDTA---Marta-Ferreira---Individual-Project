```python
import pandas as pd
likes_df = pd.read_csv('total_likes_all_parties.csv')
likes_df.head(10)
```

```python
import pandas as pd
votes_df = pd.read_excel('election_results.xlsx')
votes_df.head(10)
```

```python
# Remove 'Blanks' and 'Nulls' from the election results df
votes_df = votes_df[(votes_df['Political Party'] != 'Blanks') & (votes_df['Political Party'] != 'Nulls')]

votes_df
```

```python
# Merge the dataframes on 'Political Party'
merged_df = pd.merge(likes_df, votes_df, on='Political Party')

# Calculate the correlation
correlation = merged_df['Total Likes'].corr(merged_df['Votes'])

print("Correlation between Total Likes and Votes:", correlation)
```

```python
# Create a scatter plot of the correlation

import seaborn as sns
import matplotlib.pyplot as plt

pastel_color = sns.color_palette('pastel')[3]

plt.figure(figsize=(8, 6))
plt.scatter(merged_df['Total Likes'], merged_df['Votes'], color=pastel_color, alpha=0.7)
plt.title('Correlation between Likes and Votes')
plt.xlabel('Likes')
plt.ylabel('Votes')
plt.grid(True)
plt.show()
```
